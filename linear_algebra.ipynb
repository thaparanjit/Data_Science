### Matrix Operation

```python
A=np.array([
    [2, -1],
    [0, 3],
])    
B=np.array([
    [1, 2],
    [-3, 4],
])    
print(A+B)
print()
print(A*B)
print()
print(3*A-2*B)
```

## Determination and Inverse

```python
import numpy.linalg as LA
LA.det(A)
```

## Inverse

```python
LA.inv(A)
```

## Car questions

```python
import numpy as np
import numpy.linalg as LA

X = np.array([[4, 1.5, 100],
              [6, 2.0, 150]])
y = np.array([20, 30])

# Method 1: Using the Normal Equation: β = (X.T * X)^(-1) * X.T * y
print("Using Normal Equation: β = (X.T * X)^(-1) * X.T * y")
print()

# Compute X.T (transpose of X)
X_T = X.T
print("X.T (Transpose of X):")
print(X_T)
print()

# Compute X * X.T (2x2 matrix)
XXT = X @ X_T
print("X * X.T:")
print(XXT)
print()

# Compute (X * X.T)^(-1) (inverse)
XXT_inv = LA.inv(XXT)
print("(X * X.T)^(-1) (Inverse):")
print(XXT_inv)
print()

# Compute β = X.T * (X * X.T)^(-1) * y
beta = X_T @ XXT_inv @ y

# Predict price for Car 3: x1=8, x2=3.0, x3=200
car3_features = np.array([8, 3.0, 200])
car3_price = car3_features @ beta

print(f"Predicted price for Car 3: {car3_price:.2f}")
```
