```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing data

```python
# Load dataset

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("./Health_heart_experimental.csv")
df.head(5)
```

## Bar Chart

```python
sns.barplot(x='sex',y='DiaBP',data=df)
```

##### In the above bar chart, we can clearly see that sex doesn't affect the DiaBP i.e diastolic blood pressure.It is almost same for both male and female as average DiaBP is almost same for both.

## Box Plot

```python
sns.boxplot(df['BMI'])
```

##### In the above box plot, the median value for the BMI is between 25 and 30 i.e nearly 28-29. 

## PIE CHART

```python
 # pie chart

df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

##### From the above pie-chart, we can see taht number of female patients(0) and male patients(1) is almost equal.

## Scatter Plot

```python
sns.scatterplot(
    data=df,
    x='age',
    y='HR',
    hue='sex'
)

plt.title('Age vs HR')
plt.xlabel('Age')
plt.ylabel('HR')
plt.show()
```

##### This scatter plot is between age and heart rate. Heart rate are widely spread across all age groups, ranging approximately from 80 to 200. We can clearly see heart beat is independent of age and sex. People with different age and sex can have variable heart rate depending upon the health condition.


## Line Plot

```python
sns.lineplot(x='weightKg',y='heightCm',data=df)
```

##### The above line graph is between weight and height. This plot shows that the patients with less height have less weight. From the graph we can see that when height is increasing the weight is also increasing keeoing the direct relation between eachother.

## Histogram

```python
plt.hist(df['weightKg'])
```

##### In the above histogram, we can see that the histogram is right-skewed which indicates that the number of patients with weight below 80kg is more in number. Most patient's weight is 50-60kg. Very less patients are above 100kg which is quite good because excessively higher weight is not good for health.
